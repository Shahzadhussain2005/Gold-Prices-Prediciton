```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import csv

start_date = datetime(2024, 1, 1)   
end_date   = datetime(2025, 1, 1)   

output_file = "npr_headlines_dataset.csv"

with open(output_file, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["date", "headline", "url"])

    current_date = start_date
    total_headlines = 0

    while current_date <= end_date:
        date_str = current_date.strftime("%Y-%m-%d")
        url = f"https://www.npr.org/sections/news/archive?date={date_str}"

        print(f"Scraping {date_str} ...")
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")

        # Headline elements
        stories = soup.select("article h2.title a")
        if not stories:
            print("No headlines found for this date.")
        else:
            for story in stories:
                headline = story.get_text(strip=True)
                link = story.get("href")
                writer.writerow([date_str, headline, link])
                total_headlines += 1

        current_date += timedelta(days=1)

print(f"\nâœ… Done! Collected {total_headlines} NPR headlines into {output_file}")
```

```python
import pandas as pd
existing_df = pd.read_csv("gold_news_clean.csv", encoding="ISO-8859-1")
npr_df = pd.read_csv("npr_headlines_dataset.csv")
npr_df = npr_df.rename(columns={
    "headline": "title",
    "date": "published"
})
npr_df["source"] = "npr.org"
npr_df["published"] = pd.to_datetime(npr_df["published"]).dt.strftime("%d/%m/%Y")
npr_df = npr_df[["source", "title", "published"]]
merged_df = pd.concat([existing_df, npr_df], ignore_index=True)
merged_df.to_csv("merged_headlines_dataset.csv", index=False, encoding="utf-8")
print(f"Total rows: {len(merged_df)}")
```

```python
df = pd.read_csv("merged_headlines_dataset.csv", encoding="utf-8")
df['published'] = pd.to_datetime(df['published'], dayfirst=True, errors='coerce')
df_sorted = df.sort_values(by='published', ascending=True)
df_sorted = df_sorted.reset_index(drop=True)
df_sorted.to_csv("merged_headlines_dataset_sorted.csv", index=False, encoding="utf-8")
print(" saved as merged_headlines_dataset_sorted.csv")
```
